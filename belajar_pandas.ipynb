```python
import numpy as np
import pandas as pd

# Kita buat dataset yang agak "rusak" untuk latihan
data_rusak = {
    'Nama': ['Rian', 'Santi', 'Budi', 'Laras'],
    'Skill_Python': [80, 90, np.nan, 85], # Budi nggak ada nilainya
    'Kota': ['Jakarta', None, 'Surabaya', 'Bandung'] # Santi nggak ada kotanya
}

df_baru = pd.DataFrame(data_rusak)
print(df_baru)
```

```python
df_baru.isnull().sum()
```

```python
df_baru.isnull().sum()
```

```python
df_drop = df_baru.dropna()
print(df_drop)
```

```python
# Kita isi nilai Python yang kosong dengan rata-rata
rata_rata = df_baru['Skill_Python'].mean()
df_baru['Skill_Python'] = df_baru['Skill_Python'].fillna(rata_rata)

print(df_baru)
```

```python
# Mencari kota yang paling banyak muncul
kota_populer = df_baru['Kota'].mode()[0]
print(f"Kota yang paling sering muncul: {kota_populer}")

# Mengisi data kosong di kolom Kota dengan kota_populer
df_baru['Kota'] = df_baru['Kota'].fillna(kota_populer)

print("\n--- Data Setelah Dibersihkan ---")
print(df_baru)
```

```python
# Mencari pelamar jago Python yang ada di Bandung
filter_canggih = df_baru[(df_baru['Skill_Python'] > 80) & (df_baru['Kota'] == 'Bandung')]

print("Hasil Filter:")
print(filter_canggih)
```

```python
# Skill Python-nya di atas 85 OR (Atau) berasal dari kota Jakarta.
filter_canggih = df_baru[(df_baru['Skill_Python'] > 85) | (df_baru['Kota'] == 'Jakarta')]

print("Hasil Filter:")
print(filter_canggih)
```

```python
# Mencari pelamar: Skill_Python > 80 DAN (TIDAK tinggal di Jakarta)
filter_canggih = df_baru[(df_baru['Skill_Python'] > 80) & ~(df_baru['Kota'] == 'Jakarta')]

print("Hasil Filter:")
print(filter_canggih)
```

```python
import matplotlib.pyplot as plt

# Membuat grafik batang
plt.bar(df_baru['Nama'], df_baru['Skill_Python'], color='skyblue')

# Memberi label
plt.title('Perbandingan Skill Python Pelamar')
plt.xlabel('Nama Pelamar')
plt.ylabel('Skor Python')

# Menampilkan grafik
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mengatur tema agar lebih estetik
sns.set_theme(style="whitegrid")

# Membuat grafik batang dengan pemisahan warna berdasarkan Kota
plt.figure(figsize=(10, 6))
grafik = sns.barplot(x='Nama', y='Skill_Python', hue='Kota', data=df_baru,)

# Menambah label dan judul
plt.title('Skill Python Pelamar Berdasarkan Kota', fontsize=16)
plt.xlabel('Nama Pelamar', fontsize=12)
plt.ylabel('Skor Python', fontsize=12)

# Menampilkan grafik
plt.show()
```
